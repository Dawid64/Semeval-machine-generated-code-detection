```python
from tree_sitter import Parser, Language
import tree_sitter_python
```

```python
parser = Parser(Language(tree_sitter_python.language()))

code = """
def example(arg1):
    print('Hello', end='')
    print('World!')
    print(f'{arg1}')
"""

tree = parser.parse(bytes(code, 'utf8'))
print(type(tree))
```

```python
print(tree.root_node)
print(tree.root_node.children[0])
```

```python
tree.root_node.children
for i in tree.root_node.children[0].children:
    print(i.type)
    print(code[i.start_byte: i.end_byte])
    print()
```

```python
block = tree.root_node.children[0].children[4]
block
```

```python
for i in block.children:
    print(i.type)
    print(code[i.start_byte: i.end_byte])
    print()
```
